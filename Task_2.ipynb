```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "Dataset\\train.csv"
df = pd.read_csv(file_path)
```

### Fig 01:Survival Counts

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, color='lightblue')
plt.title("Survival Distribution")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

### Fig 02: Gender vs Survival

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

### Fig 03: Passenger Class vs Survival

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set1')
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

### Fig 04: Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True, bins=30, color='lightblue')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Fig 05: Age vs Survival

```python
plt.figure(figsize=(8,5))
ax = sns.boxplot(x='Survived', y='Age', data=df)
colors = sns.color_palette('coolwarm', n_colors=df['Survived'].nunique())
for i, artist in enumerate(ax.artists):
    artist.set_facecolor(colors[i])
    artist.set_edgecolor('k')
    artist.set_alpha(0.7)
plt.title("Age vs Survival")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```

### Fig 06: Fare vs Survival

```python
plt.figure(figsize=(8,5))
ax = sns.boxplot(x='Survived', y='Fare', data=df)
colors = sns.color_palette('coolwarm', n_colors=df['Survived'].nunique())
for i, artist in enumerate(ax.artists):
    artist.set_facecolor(colors[i])
    artist.set_edgecolor('k')
    artist.set_alpha(0.7)
plt.title("Fare vs Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

### Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues', linewidths=0.5)
```
